```python
moon = open('/content/drive/MyDrive/speech_moon.txt', encoding ='UTF-8').read()
moon
```

```python
import re
moon = re.sub('[^가-힣]', ' ', moon)
moon
```

```python
!pip install konlpy

import konlpy
```

```python
hannanum = konlpy.tag.Hannanum()

hannanum.nouns("대한민국의 영토는 한반도와 그 부속도서로 한다")
```

```python
nouns = hannanum.nouns(moon)

import pandas as pd
df_word = pd.DataFrame({'word' : nouns})
df_word
```

```python
#단어 빈도표 만들기

df_word['count'] = df_word['word'].str.len()
df_word
```

```python
# 두 글자 이상 단어만 남기기

df_word = df_word.query('count >= 2')
df_word.sort_values('count')
```

```python
# 단어 빈도 구하기
df_word = df_word.groupby('word', as_index=False)\
.agg(n=('word', 'count'))\
.sort_values('n', ascending=False)

df_word
```

```python
top20 = df_word.head(20)
top20
```

```python
!pip install wordcloud  # wordcloud 라이브러리 설치
from wordcloud import WordCloud  # wordcloud 라이브러리 불러오기
```

```python
# 단어와 빈도를 담은 딕셔너리 만들기

dic_word = df_word.set_index('word').to_dict()['n']
dic_word
```

```python
# 폰트 설정
font = '/content/drive/MyDrive/DoHyeon-Regular.ttf'

# 워드 클라우드 만들기
wc = WordCloud(random_state = 1234,
               font_path = font,
               width = 400,
               height = 400,
               background_color = 'white')

#워드 클라우드 만들기
img_wordcloud = wc.generate_from_frequencies(dic_word)

#워드 클라우드 출력하기
plt.figure(figsize=(7,7))
plt.axis('off')  # 테두리선 없애기
plt.imshow(img_wordcloud);
```

```python
# 워크 클라우드 모양 바꾸기

import PIL
icon = PIL.Image.open('/content/drive/MyDrive/cloud.png')
```

```python
img = PIL.Image.new('RGB', icon.size, (255, 255, 255))
img.paste(icon, icon)
img = np.array(img)
```

```python
wc = WordCloud(random_state = 1234,
               font_path = font,
               width = 400,
               height = 400,
               background_color = 'white',
               mask = img)

# 워드 클라우드 만들기
img_wordcloud = wc.generate_from_frequencies(dic_word)

# 출력하기
plt.figure(figsize=(7,7))
plt.axis('off')
plt.imshow(img_wordcloud)
```
