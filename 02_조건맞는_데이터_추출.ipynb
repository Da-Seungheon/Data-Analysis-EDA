# 2. 조건에 맞는 데이터만 추출

```python
import pandas as pd
import numpy as np
```

```python
df = pd.DataFrame({
    'nclass' : [2, 1, 2, 1, 1, 2],
    'english' : [98, 50, 80, 69, 89, 65],
    'science' : [50, 50, 67, 78, 44, 99]
})
df
```

### 1반 학생들의 데이터만 추출

```python
df.query('nclass == 1')
```

### 1반이 아닌 학생들만 추출

```python
df.query('nclass != 1')
```

## 초과, 미만, 이상, 이하 조건 걸기

### 영어 점수가 80점 초과한 경우

```python
df.query('english > 80')
```

### 1반이면서 과학점수가 70점 이상인 학생

```python
df.query('nclass == 1 & science > 70')
```

### 영어점수가 70점 이상이거나 과학점수가 60점 이상인 학생

```python
df.query('english >= 70 | science >= 80')
```

### in과 리스트를 이용하여 원하는 값 출력

```python
# 1, 2반에 해당하면 출력
df.query('nclass in [1, 2]')
```

# 추출한 행으로 데이터 만들기

```python
df1 = df.query('nclass == 1')
df2 = df.query('nclass == 2')

print('1반 영어 평균점수:', df1['english'].mean())
print('2반 영어 평균 점수:', df2['english'].mean())
```

# 문자 변수를 이용해 조건에 맞는 행 추출하기

```python
df = pd.DataFrame({
    'sex' : ['F', 'M', 'F', 'M', 'M', 'F'],
    'country' : ['korea', 'korea', 'japan', 'usa', 'sua', 'japan']
})

df
```

```python
# 성별이 여자이면서 나라가 한국인 사람 추출
df.query('sex == "F" & country == "korea"')
```

```python
# 여러 변수를 제거할려면 리스토로 진행
df.drop(columns='sex')
```

```python
df = pd.DataFrame({
    'nclass' : [2, 1, 2, 1, 1, 2],
    'english' : [98, 50, 80, 69, 89, 65],
    'science' : [50, 50, 67, 78, 44, 99]
})
df
```

```python
# 1반학생들의 영어점수를 추출
df.query('nclass == 1')\
 ['english']
```
